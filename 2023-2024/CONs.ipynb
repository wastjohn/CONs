```python
# calculating ISM masses for IC860
import pandas as pd
import math

# observed filters of ratioed images
l1 = 606
l2 = 814

# constants and known values
D = 29.9E6  # distance to galaxy [pc]
C = 2.05E21
m_h = 2.2E-24 # mass of hydrogen accounting for helium abundance


# taken from DS9
contour_levels = [0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5,] #1.6, 1.6]
area = [0.0141277, 0.073778, 1.06586, 2.22433, 3.75169, 7.34797, 15.6708, 82.6094,]# 5.57416, 21.7614]
median = [0.712504, 0.83612, 0.942991, 0.996487, 1.06182, 1.18652, 1.29161, 1.40902,]# 1.64068, 1.63771]
background_median = 1.641

# adding data to one row
data = []
for i in range(len(contour_levels)):
    row = [contour_levels[i], area[i], median[i], background_median]
    data.append(row)

# creating the data frame
df = pd.DataFrame(data, columns=["Contour Levels", "Total Contour Area (arcsec^2)", "Median", "Background Median"])

# populating values in dataframe
tau_v = []
area_cm2 = []
N = []
surf_den = []
M = []
for i in range(len(contour_levels)):
    # optical depth tau_v
    tau_v_i = (math.log(median[i]/background_median))/(-(l1/555)**-0.7 + (l2/555)**-0.7)
    tau_v.append(tau_v_i)

    # converting the area to cm^2
    # area_cm2_i = (D/206265)**2 * (area[i] - sum(area[:i])) * (3.085E18)**2  # not sure if I calculated masses correctly
    try:
        area_cm2_i = (D/206265)**2 * (area[i] - area[:i][-1]) * (3.085E18)**2  # TODO: check if this is calculated correctly (it might be subtracting off duplicates)
    except:
        area_cm2_i = (D/206265)**2 * (area[i] - sum(area[:i])) * (3.085E18)**2  # TODO: check if this is calculated correctly (it might be subtracting off duplicates)

    area_cm2.append(area_cm2_i)

    # column density N
    N_i = C * (tau_v[i])
    N.append(N_i)

    # surface density
    surf_den_i = N[i] * m_h
    surf_den.append(surf_den_i)

    # dust mass in g
    m_i = surf_den[i] * area_cm2[i]
    M.append(m_i)

# adding variables to the dataframe
df["\u03C4_v"] = tau_v
df["Effective Area [cm^2]"] = area_cm2
df["Column Density N [atoms/cm^2]"] = N
df["Surface Density \u03A3 [g/cm^2]"] = surf_den
df["ISM Mass [g]"] = M
df["ISM Mass [solar masses]"] = [x / 1.99E33 for x in M]

# total dust mass
total_ISM_mass = sum(df["ISM Mass [g]"])
total_ISM_solar_mass = sum(df["ISM Mass [solar masses]"])
print(f"Total ISM Mass = {total_ISM_mass} g = {total_ISM_solar_mass} solar masses ")

# displaying the dataframe
df
```
