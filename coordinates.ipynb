```python
from astroquery.ipac.ned import Ned
from astroquery.ipac.irsa import Irsa
from astroquery.vizier import Vizier
from astropy.coordinates import SkyCoord
import astroquery
import pandas as pd
import numpy as np
from astropy import units as u
```

```python
# read in data
df = pd.read_csv('data/CONTargets.csv')
names = df['Name'].values
```

```python
# convert hms dms to degrees
ra = df['RA_J2000'].values
dec = df['Dec_J2000'].values
for i in range(len(df)):
    try:
        if ":" in ra[i] or ":" in dec[i]:
            sc = SkyCoord(ra[i], dec[i], unit=("hourangle", "deg"))
            df.loc[i, "RA_J2000"] = sc.ra.deg
            df.loc[i, "Dec_J2000"] = sc.dec.deg
    except:
        pass

# add search results to missing data
df.to_csv('data/CONTargetList_filled.csv', index=False)
```

```python
nishimura_ned = Ned.query_refcode('2024A&A...686A..48N')["Object Name", "RA", "DEC", "Redshift"]  # Nishimura et al
u_ned = Ned.query_refcode('2012ApJS..203....9U')["Object Name", "RA", "DEC", "Redshift"]  # U et al

u = u_ned.to_pandas()
u.tail(n=60)
# I just need the redshifts from u et al
# and the coordinates from nishimura and johnstone
```

```python
df2 = pd.read_csv('data/CONTargetList_filled.csv')
ra, dec = df2['RA_J2000'].values, df2['Dec_J2000'].values


file = open("data/CONTargetList_NoLeroyMass.csv", "w")
file.write("index,name,ra,dec\n")
nofound = 0
for i in range(len(df2)):
    sc = SkyCoord(ra[i], dec[i], unit=("deg", "deg"))
    result = Vizier.query_region(sc,
                                radius=15*u.arcsec,
                                catalog='J/ApJS/244/24/table4')
    if len(result) == 0:
        print(df2['Name'][i], "not found")
        file.write(f"{nofound+1},{df2['Name'][i]},{df2['RA_J2000'][i]},{df2['Dec_J2000'][i]}\n")
        nofound += 1
    elif len(result) == 1 and len(result[0]) > 1:
        print(df2['Name'][i], "multiple sources found")
    else:
        print(result[0]["logM_"])

file.close()

print(f"{nofound} targets not found ({nofound/len(df2)*100}%)")
```
